 **Задача 1**

 Из колоды в 52 карты извлекаются случайным образом 4 карты.

 a) Найти вероятность того, что все карты– крести.

 б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз

```python
import math

all_cards = 52
cristen = 13
aces = 4

all_combinations = math.comb(all_cards, 4)

# a)
cristen_comb = math.comb(cristen, 4)
cristen_probability = cristen_comb / all_combinations
print(f'Вероятность того, что все карты крести: {cristen_probability * 100: .2f}%')
# b)
no_aces_comb = math.comb(all_cards - aces, 4)
no_aces_probability = no_aces_comb / all_combinations
one_ace_probability = 1 - no_aces_probability
print(f'Вероятность того, что среди 4х карт останется хотя бы один туз: {one_ace_probability * 100: .2f}%')
```

 **Задача 2.**

 На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с
 цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно.
 Какова вероятность того, что человек, не знающий код, откроет дверь с первой
 попытки?

```python
import math

buttons = 10
code = 3

total_combinations = math.comb(buttons, code)
result_probability = 1 / total_combinations

print(f'Вероятность того, что человек откроет дверь с первой попытки: {result_probability * 100: .2f}%')
```

**Задача 3**

 В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом
 извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
import math

details = 15
painted = 9
taken = 3

all_taken_comb = math.comb(details, taken)
painted_taken_comb = math.comb(painted, taken)

painted_taken_probability = painted_taken_comb / all_taken_comb
print(f'Вероятность того, что три взятые детали будут окрашены: {painted_taken_probability * 100: .2f}%')
```

**Задача 4.**

В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
import math

tickets = 100
win_tickets = 2
purchased = 2

all_purchased_comb = math.comb(tickets, purchased)
win_purchased_comb = math.comb(win_tickets, purchased)
win_purchased_probability = win_purchased_comb / all_purchased_comb
print(f'Вероятность того, что 2 приобретенных билета окажутся выигрышными: {win_purchased_probability * 100: .2f}%')
```
