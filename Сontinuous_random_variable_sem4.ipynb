**Задача 4.**
 Веса изделий следуют нормальному распределению с дисперсией 9 и средним
арифметическим 50.  
 Веса ниже 45 г считаются браком. Какая доля изделий бракованная?


mu = 50  
σ^2 = 9  
X < 45  - брак
Z = (45-50) / sqrt(9) = -5/3 = -1,67

p = 4,75%  
брак - 0,0475

```python
import math

D = 9
mu = 50
X = 44

std = math.sqrt(D)
print(f'{std=}')
Xstand = (X - mu) / std
print(f'{Xstand=}, 2.3% изделий будут браком')
```

```python
import math
from scipy.stats import norm

D = 9
mu = 50
X = 45

std = math.sqrt(D)
print(f'{std=}')
Xstand = (X - mu) / std
print(f'{Xstand= :.2f}')
p_value = norm.sf(abs(Xstand))
print(f'{p_value=}')
```

**Задача 5.**  
Дана генеральная совокупности из задачи 3.  
 Найдите долю изделий, с весом:  
 1) выше 56 г  
 2) весом меньше 55 г  
 3) Весом не больше 50 г  
 4) Весом не больше 55 г, но не меньше 45 г  
 5) Весом  не меньше 48 г  
 6) Какая доля продукции весит меньше 54 г  
 7) Какая доля продукции весит больше 46 г  
 8) Какая доля продукции весит больше 52 г  
 9) Какая доля продукции имеет вест от 44 до 54 г?




```python
import math
from scipy.stats import norm

D = 9
mu = 50
std = math.sqrt(D)

# 1) доля изделий выше 56
prob_1 = 1 - norm.cdf(56, mu, std)
print(f'1) Доля изделий выше 56 г. {prob_1:.3f}')

# 2) доля изделий меньше 55г
prob_2 = norm.cdf(55, mu, std)
print(f'2) Доля изделий меньше 55 г. {prob_2:.3f}')

# 3) доля изделий не больше 50г
prob_3 = 1 - norm.cdf(50, mu, std)
print(f'3) Доля изделий не больше 50 г. {prob_3:.3f}')

# 4) доля изделий не больше 55 г, но не меньше 45 г
prob_4 = norm.cdf(55, mu, std) - norm.cdf(45, mu, std)
print(f'4) Доля изделий не больше 55 г, но не меньше 45 г {prob_4:.3f}')

# 5) Доля изделий не меньше 48 г
prob_5 = 1 - norm.cdf(48, mu, std)
print(f'5) Доля изделий не меньше 48 г {prob_5: .3f}')

# 6) Какая доля продукции весит меньше 54 г
prob_6 = norm.cdf(54, mu, std)
print(f'6) Доля изделий меньше 54 г {prob_6: .3f}')

# 7) Какая доля продукции весит больше 46 г
prob_7 = 1 - norm.cdf(46, mu, std)
print(f'7) Доля изделий больше 46 г {prob_7: .3f}')

# 8) Какая доля продукции весит больше 52 г
prob_8 = norm.cdf(52, mu, std)
print(f'8) Доля изделий больше 52 г {prob_8: .3f}')

# 9) Какая доля продукции имеет вест от 44 до 54 г?
prob_9 = norm.cdf(54, mu, std) - norm.cdf(44, mu, std)
print(f'9) Доля изделий весом от 44 до 54 г {prob_9: .3f}')
```

 **Задача 2.**  
 О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?  
 Если да, найдите ее.

```python
import math

D_X = 0.2
a = 0.5

# D = (b-a)^2 / 12
# (b-a)^2 = 12*D
# b-a = sqrt(12*D)
b = math.sqrt(12*D_X) + a
M_X = (a+b)/2

print(f'Правая величина: {b= :.2f}, среднее значение: {M_X= :.2f}')
```

 **Задача 4.**  
 Рост взрослого населения города X имеет нормальное распределение.  
 Причем,
 средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.   
 Какова вероятность того, что случайным образом выбранный взрослый человек
 имеет рост:  
 а) больше 182 см  
 б) больше 190 см  
 в) от 166 см до 190 см  
 г) от 166 см до 182 см  
 д) от 158 см до 190см  
 е) не выше 150 см или не ниже 190 см  
 ё) не выше 150 см или не ниже 198 см  
 ж) ниже 166 см.

```python
from scipy.stats import norm

mu = 174
std = 8

print(f'а) Рост больше 182: {(1 - norm.cdf(182, mu, std)): .2f}')

print(f'б) Рост больше 190 см: {(1 - norm.cdf(190, mu, std)) : .2f}')

print(f'в) Рост от 166 см до 190 см: {(norm.cdf(190, mu, std) - norm.cdf(166, mu, std)): .2f}')

print(f'г) Рост от 166 см до 182 см: {(norm.cdf(182, mu, std) - norm.cdf(166, mu, std)): .2f}')

print(f'д) Рост от 158 см до 190см: {(norm.cdf(190, mu, std) - norm.cdf(158, mu, std)): .2f}')

print(f'е) Рост не выше 150 см или не ниже 190 см: {(norm.cdf(150, mu, std) + (1 - norm.cdf(190, mu, std))): .2f}')

print(f'ё) Рост не выше 150 см или не ниже 198 см: {(norm.cdf(150, mu, std) + (1 - norm.cdf(198, mu, std))): .2f}')

print(f'ж) Рост ниже 166 см.: {norm.cdf(166, mu, std): .2f}')
```

 **Задача5.**  
 Насколько сигм (средних квадратичных отклонений) отклоняется рост человека,
 равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?


```python
import math

M_X = 178
D_X = 25
height = 190

std = math.sqrt(D_X)
z_score = (height - M_X) / std

print(f'Рост 190 см отклоняется от среднего на {z_score: .2f} сигм')
```
