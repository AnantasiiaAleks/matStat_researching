**Задача 2.**  
Исследовалось влияние препарата на уровень давления пациентов. Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут. Есть ли статистически значимые различия?  
1-е измерение до приема препарата: 150, 160, 165, 145, 155  
2-е измерение через 10 минут: 140, 155, 150, 130, 135  
3-е измерение через 30 минут: 130, 130, 120, 130, 125

```python
import numpy as np
import scipy.stats as sps

x1 = np.array([150, 160, 165, 145, 155])
x2 = np.array([140, 155, 150, 130, 135])
x3 = np.array([130, 130, 120, 130, 125])

alpha = 0.05

u_statistic, p_value = sps.friedmanchisquare(x1, x2, x3)

if p_value > alpha:
  print(f'{p_value= : .3f} > {alpha= }, статистически значимых различий не наблюдается, принимаем нулевую гипотезу.')
else:
  print(f'{p_value= : .3f} < {alpha= }, наблюдаются статистически значимые различия, есть основания отвергнуть нулевую гипотезу и принять альтернативную.')
```

**Задача 4.**  
Даны 3 группы учеников плавания.

```python
import numpy as np
import scipy.stats as sps

A = np.array([56, 60, 62, 55, 71, 67, 59, 58, 64, 67])
B = np.array([57, 58, 69, 48, 72, 70, 68, 71, 50, 53])
C = np.array([57, 67, 49, 48, 47, 55, 66, 51, 54])

alpha = 0.05

u_statistic, p_value = sps.kruskal(A, B, C)

if p_value > alpha:
  print(f'{p_value= : .3f} > {alpha= }, статистически значимых различий не наблюдается, принимаем нулевую гипотезу.')
else:
  print(f'{p_value= : .3f} < {alpha= }, наблюдаются статистически значимые различия, есть основания отвергнуть нулевую гипотезу и принять альтернативную.')
```

 Если ANOVA показывает значимые различия, рекомендуется провести пост-хок тесты
 (например, тест Тьюки), чтобы определить, какие именно группы отличаются друг от
 друга.

```python
import numpy as np
import scipy.stats as stats
 # Данные
gr1 = [56, 60, 62, 55, 71, 67, 59, 58, 64, 67]
gr2 = [57, 58, 69, 48, 72, 70, 68, 71, 50, 53]
gr3 = [57, 67, 49, 48, 47, 55, 66, 51, 54]
alpha=0.05

 # ANOVA (анализ вариаций) используется для сравнения средних значений трех и более групп. Этот метод позволяет определить, есть ли статистически значимые различия между группами.

 # Перед использованием ANOVA важно убедиться, что данные в группах распределены нормально. Если нормальность не соблюдается, может потребоваться использование непараметрических тестов, таких как тест Краскела-Уоллиса.

stat_gr1, p_gr1 = stats.shapiro(gr1)   # p-значение 0.7333586431215993 > alpha, нормальность подтверждена
stat_gr2, p_gr2 = stats.shapiro(gr2)   # p-значение 0.09523341958284404 > alpha, нормальность подтверждена
stat_gr3, p_gr3 = stats.shapiro(gr3)   # p-значение 0.16166250659063675 > alpha, нормальность подтверждена

# ANOVA предполагает, что дисперсии внутри групп равны. Если дисперсии существенно отличаются, результаты теста могут быть недостоверными. Для проверки этого предположения можно использовать тест Левена.

levene_stat, p_value_levene = sps.levene(gr1, gr2, gr3)    # p_value = 0.05050645982348729 -> дисперсии существенно не отличаются

f_statistic, p_value = stats.f_oneway(gr1, gr2, gr3)
print(f"F-статистика: {f_statistic}")     # 2.580607609925374
print(f"p-значение: {p_value}")           # 0.0949875108784058

# F-статистика указывает на соотношение межгрупповой дисперсии к внутригрупповой. Если p-значение меньше уровня значимости (обычно 0.05), это свидетельствует о наличии статистически значимых различий между группами.
if p_value < alpha:
  print("Есть статистически значимые различия между группами.")
else:
  print("Нет статистически значимых различий между группами.")
```

**Задача 5.**  
Заявляется, что партия изготавливается со средним арифметическим 2,5 см.
Проверить данную гипотезу, если известно, что размеры изделий подчинены
нормальному закону распределения.  
Объем выборки 10, уровень статистической значимости 5%  
2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

```python
# H0: партия изготавливается со средним арифметическим 2,5
# H1: среднее арифметическое партии отличается от 2,5

import numpy as np
import scipy.stats as sps

x = np.array([2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34])
n = 10
alpha = 0.05
mu_0 = 2.5

x_mean = np.mean(x)
std_x = np.std(x, ddof=1)
t_stat = (x_mean - mu_0) / (std_x / np.sqrt(n))      # 0.5630613661802959

df = n - 1        # 9
t_crit = sps.t.ppf(1-alpha/2, df)                    # 2.2621571628540993

# p-значение
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Результаты
print(f"Среднее выборки: {x_mean: .2f}")
print(f"t-статистика: {t_stat: .3f}")
print(f"p-значение: {p_value: .4f}")

if abs(t_stat) > t_crit:
  print("Отвергаем нулевую гипотезу.")
else:
  print("Нет оснований отвергать нулевую гипотезу.")

if p_value < alpha:
    print("Отвергаем нулевую гипотезу.")
else:
    print("Нет оснований отвергать нулевую гипотезу.")

```
