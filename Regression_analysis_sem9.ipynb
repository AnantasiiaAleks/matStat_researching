````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
from sklearn.linear_model import LinearRegression
from math import comb
````

````python
X1= np.array([30, 30, 40, 40])
Y1= np.array([37, 47, 50, 60])

model1 = LinearRegression()
X1 = X1.reshape(-1, 1)
regres1 = model1.fit(X1, Y1)
print(f'b0: {regres1.intercept_=}')
print(f'b1: {regres1.coef_=}')

y_pred1 = model1.predict(X1)
print(y_pred1)

print(f'Коэффициент детерминации R^2: {regres1.score(X1, Y1)}')
````

````python
plt.scatter(X1, Y1)
plt.plot(X1, y_pred1)
plt.title('r^2= 0.6282527881040892')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

````python
tickets = 10
win_tickets = 2
lose_tickets = 8
taken = 5
taken_win = 2
taken_lose = 5


C_taken_from_all = comb(tickets, taken)     # 252
C_taken_lose = comb(lose_tickets, taken_lose)   # 56
C_taken_win = comb(win_tickets, taken_win)    # 1
C_good = C_taken_win * C_taken_lose           # 56
P = C_good / C_taken_from_all
print(f'Вероятность составляет: {P*100 : .2f}%')
````

Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):  
zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]  
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].  
Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp- признак), а за Y - значения скорингового балла (то есть, ks- целевая переменная). Произвести расчет как с использованием intercept, так и без.

 Подсказка № 1  
 Ковариация между двумя переменными помогает определить, насколько они
 изменяются вместе. В Python ковариацию между zp и ks можно найти через формулу:  
 Cov(X,Y)= ∑(𝑋−𝑋¯)⋅(𝑌−𝑌¯) / 𝑛−1
 , где X и Y —переменные, а Xˉ и Yˉ — их средние значения.

 Подсказка № 2  
 Рассчитать коэффициент наклона без intercept. Для получения коэффициента наклона
 (b) без intercept используйте формулу:  
 b= 𝐶𝑜𝑣(𝑋,𝑌) /  𝑉𝑎𝑟(𝑋)
Дисперсию (Var) можно найти как
 ∑(𝑋−𝑋¯) ** 2 /  𝑛−1  
В Python это делается через np.sum и np.mean.

 Подсказка № 3  
 При расчете коэффициентов с intercept, формула для intercept (a) следующая:
 a=Yˉ−b⋅Xˉ. Это значение добавляет смещение к регрессионной линии, чтобы она
 проходила через средние значения переменных.

````python
import numpy as np
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

zp_mean = np.mean(zp)
ks_mean = np.mean(ks)

n = len(zp)             # = len(ks)

# Ковариация и дисперсия
Cov_zp_ks = np.sum((zp - zp_mean) * (ks - ks_mean)) / (n - 1)       # 10175.37777777778
Var_zp = np.sum((zp - zp_mean)**2) / (n - 1)                        # 3882.9333333333334

# Без intercept
b = Cov_zp_ks / Var_zp                                              # 2.620538882402766

# Интерсепт
a = ks_mean - b * zp_mean                                           # 444.1773573243595
````

 **Задача 2.**  
Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

Подсказка № 1  
Градиентный спуск — это метод оптимизации, который помогает находить значения параметров, минимизирующие функцию потерь. В этом случае, функция потерь — это среднеквадратичная ошибка. Градиентный спуск обновляет коэффициенты на основе градиента функции потерь относительно этих коэффициентов.  
Подсказка № 2  
Градиент для линейной регрессии с одной переменной (без intercept) вычисляется как  
1/𝑛 * ∑(y−(b⋅x))⋅x, где b — текущий коэффициент наклона.  
Подсказка № 3  
Скорость обучения определяет, насколько большие шаги делаются при обновлении коэффициента наклона. Выберите скорость обучения так, чтобы градиентный спуск сходился, но не прыгал вокруг оптимального значения. Если скорость обучения слишком велика, алгоритм может не сойтись.  
Подсказка № 4  
Количество итераций должно быть достаточным, чтобы алгоритм успел сойтись. При небольшом числе итераций может не хватить времени для нахождения оптимального решения. Следите за значением MSE и,если необходимо, увеличьте число итераций.  
Подсказка № 5  
Среднеквадратичная ошибка (MSE) вычисляется как  
1/n * ∑(𝑦−𝑦^)**2 ,где 𝑦^ предсказанные значения.  
Это значение помогает отслеживать, насколько хорошо модель подгоняет данные на каждом шаге градиентного спуска.  
Подсказк а№ 6  
На каждом шаге градиентного спуска важно выводить текущие значения коэффициента и MSE. Это позволит увидеть, как меняется ошибка и как быстро сходится модель.

````python
import numpy as np
from math import e

zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

alpha = 0.000001
b = 0.1
n = len(zp)             # = len(ks)
iterations = 1000

def mse(b, x=zp, y=ks, n=10):
  return (np.sum(b* x - y)**2) / n

for i in range(iterations):
  b -= alpha * (2/n) * np.sum((b * zp - ks) * zp)
  print(f'{i}: {b=}, mse= {mse(b)}')

print(f'Коэффициент склонности: {b: .3f}')          # Коэффициент склонности:  5.890
````

Подсказка № 6  
Начальные значения коэффициентов (например, w и b) могутбыть любыми, но их правильная инициализация важна для стабильности обучения. Обычно инициализация производится нулями или малыми случайными числами.

````python
import numpy as np

zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

alpha = 0.000001
w = 0
b = 0

n = len(zp)             # = len(ks)
iterations = 5000

def mse(w, x=zp, y=ks, n=10):
  return np.mean((w * x - y) ** 2)

for i in range(iterations):
  y_pred = w * zp + b                                   # формула линейной регрессии для каждой предсказанной точки
  w -= alpha * (2 / n) * np.sum((y_pred - ks) * zp)     # формула из подсказок
  b -= alpha * (2 / n) * np.sum(y_pred - ks)            # формула из подсказок
  # y_errors = y_pred - ks                                # Ошибки в точке, разница предсказанной и фактической
  # w -= alpha * (2 / n) * np.dot(zp, y_errors)           # вариант из решения
  # b -= alpha * (2 / n) * np.sum(y_errors)               № вариант из решения
  if i % 100 == 0:
    print(f'{i}: {w=}, {b=}, mse= {mse(w)}')


print(f'Коэффициент склонности: {w: .3f}, интерсепт: {b: .3f}')

# 1) Коэффициент склонности:  5.881, интерсепт:  1.168
# 2) Коэффициент склонности:  5.881, интерсепт:  1.168
````
