**Задача 1.**  
На препарате А положительный результат лечения наблюдается у 17 из 32 пациентов, а на препарате В у 9 из 22. Построить 95% доверительный интервал для разности долей.  
**Задача 2.**  
Обнаружены ли статистически значимые различия?

```python
import numpy as np

n1 = 32
m1 = 17

n2 = 22
m2 = 9

p1 = m1 / n1
p2 = m2 / n2

p = (m1 + m2) / (n1 + n2)

z = (abs(p1 - p2) - 0.5 * (1/n1 + 1/n2)) / np.sqrt(p * (1 - p) * (1/n1 + 1/n2))
z_tab = 1.96  # табличное значение z-критерия для альфа 5%

if z > z_tab:
  print(f'{z=}, {z_tab=}. Расчетный критерий попадает в область принятия альтернативной гипотезы: статистически значимые различия обнаружены')
else:
  print(f'{z=}, {z_tab=}. Расчетное значение Z не попало в критическую область. Оснований для отверждения нулевой гипотезы нет. Статистически значимых различий не обнаружено.')


delta = p1 - p2
s_delta = np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

lower = delta - z_tab * s_delta
upper = delta + z_tab * s_delta

print(f'Доверительный интервал: [{lower: .3f} : {upper: .3f}]')
```

**Задача 3.**  
Было проведено исследование научных статей на количество авторов в разные годы.
Построить 90% и 95% интервалы
Снимок экрана 2025-03-05 083710.png


```python
import numpy as np
import scipy.stats as sps

alpha1 = 0.1
alpha2 = 0.05


def interval(n, x, std, alpha):
  z_crit = sps.norm.ppf(1 - alpha / 2)
  l = x - z_crit * (std / np.sqrt(n))
  u = x + z_crit * (std / np.sqrt(n))
  return l, u


# Для 1946 года
l_1946_1, u_1946_1 = interval(151, 2, 1.4, alpha1)
l_1946_2, u_1946_2 = interval(151, 2, 1.4, alpha2)
print(f'Интервал для 90% для 1946 года: [{l_1946_1: .3f} : {u_1946_1: .3f}]')
print(f'Интервал для 95% для 1946 года: [{l_1946_2: .3f} : {u_1946_2: .3f}]')

# Для 1956 года
l_1956_1, u_1956_1 = interval(149, 2.3, 1.6, alpha1)
l_1956_2, u_1956_2 = interval(149, 2.3, 1.6, alpha2)
print(f'Интервал для 90% для 1956 года: [{l_1956_1: .3f} : {u_1956_1: .3f}]')
print(f'Интервал для 95% для 1956 года: [{l_1956_2: .3f} : {u_1956_2: .3f}]')
# Для 1966 года
l_1966_1, u_1966_1 = interval(157, 2.8, 1.2, alpha1)
l_1966_2, u_1966_2 = interval(155, 2.8, 1.2, alpha2)
print(f'Интервал для 90% для 1966 года: [{l_1966_1: .3f} : {u_1966_1: .3f}]')
print(f'Интервал для 95% для 1966 года: [{l_1966_2: .3f} : {u_1966_2: .3f}]')

# Для 1976 года
l_1976_1, u_1976_1 = interval(155, 4.9, 7.3, alpha1)
l_1976_2, u_1976_2 = interval(155, 4.9, 7.3, alpha2)
print(f'Интервал для 90% для 1976 года: [{l_1976_1: .3f} : {u_1976_1: .3f}]')
print(f'Интервал для 95% для 1976 года: [{l_1976_2: .3f} : {u_1976_2: .3f}]')
```

**Задача 4**

```python
import numpy as np
import scipy.stats as sps

alpha = 0.1
D = 3.6
x_mean = 71.2
n = 100

z_crit = sps.norm.ppf(1-alpha/2)

l = x_mean - z_crit * np.sqrt(D / n)
u = x_mean + z_crit * np.sqrt(D / n)

print(f'[{l: .3f} : {u: .3f}]')
```

**Задача 5.**  
Найдите 95% доверительные интервалы для долей больных, которые не чувствовали боли при включенном и выключенном приборе.  
Можно ли по этим интервалам оценить статистическую значимость различий?  
__________________________
Прибор включен: боли нет 24, боль есть 6.  
Прибор выключен: болит нет 3, боль есть 17

```python
import numpy as np
import scipy.stats as sps

alpha = 0.05



def interval_proportion(a, b, alpha):
  z_crit = sps.norm.ppf(1-alpha/2)
  n = a + b
  p = a / n
  l = p - z_crit * np.sqrt((p * (1 - p)) / n)
  u = p + z_crit * np.sqrt((p * (1 - p)) / n)
  return l, u

# Прибор включен
on_bolinet = 24
on_bol = 6
on_l, on_u = interval_proportion(on_bolinet, on_bol, alpha)


# Прибор выключен
off_bolinet = 3
off_bol = 17
off_l, off_u = interval_proportion(off_bolinet, off_bol, alpha)

print(f'При включенном приборе интервал: [{on_l: .3f} : {on_u: .3f}]')
print(f'При выключенном приборе интервал: [{off_l: .3f} : {off_u: .3f}]')

'''
При включенном приборе интервал: [ 0.657 :  0.943]
При выключенном приборе интервал: [-0.006 :  0.306]
Поскольку доверительные интервалы не пересекаются (первый начинается с 0.657, а второй идет до 0.306), это указывает на то, что различия между показателями статистически значимы.
'''
```

**Задача 6.**  
Даны две группы студентов. В каждой по 21 человеку. Первая группа прошла недельную подготовку для сдачи тестов, другая группа приступила к тестам без подготовки.  
В 1й группе среднее время прохождения теста целиком требует 1.8 ч, во второй 2.3 ч.   
Постройте 95% интервал для разности средних арифметических, если известно, что
стандартное отклонение 1 группы составило 0.4 ч, во 2й группе -0.6ч

```python
import numpy as np
import scipy.stats as sps

n1 = n2 = 21
x1 = 1.8
x2 = 2.3
alpha = 0.05
std1 = 0.4
std2 = 0.6

delta = round(abs(x1 - x2), 2)
df = n1 + n2 - 2
t_crit = abs(sps.t.ppf(alpha/2, df))
D_all = 0.5 * (std1 ** 2 + std2 **2)
SE = np.sqrt(D_all / n1 + D_all / n2)

l = delta - t_crit * SE
u = delta + t_crit * SE

print(f'{delta= }, {df= }, {t_crit= }, {D_all= }, {SE= }')

print(f'Интервал для разности средних арифметических: [{l: .3f} : {u: .3f}]')
```

Критерий Краскела-Уоллиса — непараметрический статистический критерий, используемый для сравнения трёх и более независимых выборок по количественному или порядковому признаку. 1

Он позволяет оценить различия между несколькими выборками по степени выраженности какого-либо признака, не указывая, однако, на направление этих изменений. 2

Условия применения критерия:

не менее трёх выборок испытуемых объектов; 1
зависимая переменная должна измеряться в порядковой или непрерывной шкале; 1
наблюдения должны быть независимыми (не должно быть никаких отношений между двумя группами или внутри каждой группы); 1
не требует наличия нормального распределения. 1
Критерий Краскела-Уоллиса подходит для сравнения небольших выборок. Желательно, чтобы в каждой выборке было не менее 5 наблюдений. 1

Для расчёта критерия нужно определить нулевую и альтернативную гипотезы, вычислить величину статистики критерия (нулевой гипотезы), определить критические значения и соответствующий им уровень значимости, интерпретировать величину р и результаты. 1

Также для расчёта критерия можно использовать готовые статистические программы, в основе которых уже заложены специальные алгоритмы для расчётов. 1

```python
import scipy.stats as sps

gr1=np.array([0.5, 0.7, 1, 1.2, 1.4])
gr2= np.array([1.3, 1.45, 1.6, 1.7, 1.8])
gr3=np.array([6.2, 12.6, 13.2, 14.1, 14.2])

sps.kruskal(gr1, gr2, gr3)
# KruskalResult(statistic=12.02000000000001, pvalue=0.002454088180413905)
'''
Тест Краскала-Уоллиса использует следующие нулевые и альтернативные гипотезы:
Нулевая гипотеза (H0): медиана одинакова во всех группах. 4
Альтернативная гипотеза (Ha): медиана не одинакова во всех группах. 4
Отклонение нулевой гипотезы не указывает, какая из групп отличается. Чтобы это выяснить, необходимо проводить апостериорные (попарные) сравнения между группами
'''
# p_value низкий, есть основания отвергнуть нулевую гипотезу, выборки разные

# применяем критерий U Манна-Уитни попарно, чтобы выяснить, где же различия
sps.mannwhitneyu(gr1, gr2)  # MannwhitneyuResult(statistic=1.0, pvalue=0.015873015873015872)
sps.mannwhitneyu(gr2, gr3)  # MannwhitneyuResult(statistic=0.0, pvalue=0.007936507936507936)
sps.mannwhitneyu(gr1, gr3)  # MannwhitneyuResult(statistic=0.0, pvalue=0.007936507936507936)
# Различия есть между всеми группами, наибольшие значения - между третьей группой и остальными
```

 **Задание 1.**  
Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16.  
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M=80, а объем выборки n=256.

```python
import numpy as np
import scipy.stats as sps

std = 16
alpha = 0.05
x_mean = 80
n = 256

z_crit = sps.norm.ppf(1-alpha/2)

l = x_mean - z_crit * std / np.sqrt(n)
u = x_mean + z_crit * std / np.sqrt(n)

print(f'Доверительный интервал составляет [{l: .3f} : {u: .3f}]')
```

 **Задача 2.**  
В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1.  
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, oценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
import numpy as np
import scipy.stats as sps

x = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
alpha = 0.05

x_mean = np.mean(x)
D = np.var(x, ddof=1)
t_crit = abs(sps.t.ppf(alpha/2, len(x) - 1))

l = x_mean - t_crit * np.sqrt(D / len(x))
u = x_mean + t_crit * np.sqrt(D / len(x))

print(f'Доверительный интервал составляет [{l: .3f} : {u: .3f}]')
```

 **Задача 3.**  
 Рост дочерей 175, 167, 154, 174, 178, 148, 160, 167, 169, 170  
 Рост матерей 178, 165, 165, 173, 168, 155, 160, 164, 178, 175  
 Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.

```python
import numpy as np
import scipy.stats as sps

x = np.array([175, 167, 154, 174, 178, 148, 160, 167, 169, 170])
y = np.array([178, 165, 165, 173, 168, 155, 160, 164, 178, 175])
alpha = 0.05

x_mean = np.mean(x)
y_mean = np.mean(y)
delta = abs(x_mean - y_mean)

D1 = np.var(x, ddof=1)
D2 = np.var(y, ddof=1)
D_all = (D1 + D2) / 2

SE = np.sqrt(D_all/len(x) + D_all/len(y))

df = (len(x) + len(y)) - 2
t_crit = abs(sps.t.ppf(alpha/2, df))

l = delta - t_crit * SE
u = delta + t_crit * SE

print(f'Доверительный интервал составляет [{l: .3f} : {u: .3f}]')
```
