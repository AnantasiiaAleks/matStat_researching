В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых.  
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4.  
Какова вероятность того, что 3 мяча белые?

```python
import math

# Условия задачи
box_1_all = 8
box_1_white = 5
box_2_all = 12
box_2_white = 5
box_1_taken = 2
box_2_taken = 4
one_ball = 1

# Вычисление вероятностей

# Первый случай: из первого ящика 2 белых, из второго ящика - 1 белый.
# Вероятность двух белых мячей из первого ящика
prob_1_box_two_white = math.comb(box_1_white, box_1_taken) / math.comb(box_1_all, box_1_taken)
# Вероятность одного белого мяча из второго ящика
prob_2_box_one_white = math.comb(box_2_white, one_ball) * math.comb(box_2_all - box_2_white, box_2_taken - one_ball) / math.comb(box_2_all, box_2_taken)
first_prob = prob_1_box_two_white * prob_2_box_one_white

# Второй случай: из первого ящика 1 белый, из второго ящика - 2 белых.
# Вероятность одного белого из первого ящика
prob_1_box_one_white = math.comb(box_1_white, one_ball) * math.comb(box_1_all - box_1_white, box_1_taken - one_ball) / math.comb(box_1_all, box_1_taken)
# Вероятность двух белых мячей из второго ящика.
prob_2_box_two_white = math.comb(box_2_white, one_ball * 2) * math.comb(box_2_all - box_2_white, box_2_taken - one_ball * 2) / math.comb(box_2_all, box_2_taken)
second_prob = prob_1_box_one_white * prob_2_box_two_white

# Третий случай: из первого ящика нет белых, из второго ящика 3 белых.
# Вероятность отсутствия белых в паре из первого ящика
prob_1_box_none_white = 1 - prob_1_box_two_white - prob_1_box_one_white
# Вероятность трех белых из второго ящика
prob_2_box_three_white = math.comb(box_2_white, one_ball * 3) * math.comb(box_2_all - box_2_white, box_2_taken - one_ball * 3) / math.comb(box_2_all, box_2_taken)
third_prob = prob_1_box_none_white * prob_2_box_three_white

# Суммирование вероятностей
general_prob = first_prob + second_prob + third_prob

print(f'Вероятность, что из ящиков будут вытащены в числе прочих три белых мяча, составляет {general_prob * 100 : .2f}%')
```

 **Задача 3.**  
 На соревновании по биатлону один из трех спортсменов стреляет и попадает в
 мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго
 —0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен:  
 a) первым спортсменом  
 б) вторым спортсменом  
 в) третьим спортсменом.

```python
p1 = 0.9
p2 = 0.8
p3 = 0.6

p_total = p1 + p2 + p3

prob_1 = p1 / p_total
prob_2 = p2 / p_total
prob_3 = p3 / p_total

print('Вероятность того, что выстрел был произведен:')
print(f'a) первым спортменом - {prob_1 * 100 : .2f}%;')
print(f'б) вторым спортменом - {prob_2 * 100 : .2f}%;')
print(f'в) третьим спортменом - {prob_3 * 100 : .2f}%.')
```

**Задача 4.**  
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе.  
Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9.  
Студент сдал первую сессию. Какова вероятность, что он учится:  
a) на факультете A  
б) на факультете B  
в) на факультете C?

```python
# Вероятность сдачи сессии для каждого из факультетов
pA = 0.8
pB = 0.7
pC = 0.9

x = 1
fac_a = x
fac_b = x
fac_c = x+x
all_facs = fac_a + fac_b + fac_c # все cтуденты

# Пропорции студентов по факультетам
prob_students_A = fac_a / all_facs
prob_students_B = fac_b / all_facs
prob_students_C = fac_c / all_facs

# Сдача сессий по факультетам
prob_session_A = prob_students_A * pA
prob_session_B = prob_students_B * pB
prob_session_C = prob_students_C * pC

# Общая вероятность сдачи сессии
general_session = prob_session_A + prob_session_B + prob_session_C

# Для каждого факультета
prob_student_session_A = prob_session_A / general_session
prob_student_session_B = prob_session_B / general_session
prob_student_session_C = prob_session_C / general_session

print('Вероятность, что студент, сдавший сессию, учится на:')
print(f'а) факультете А: {prob_student_session_A * 100 : .2f}%')
print(f'б) факультете B: {prob_student_session_B * 100 : .2f}%')
print(f'в) факультете C: {prob_student_session_C * 100 : .2f}%')
```

**Задача 5.**  
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25.  
Какова вероятность того, что в первый месяц выйдут из строя:  
а) все детали  
б) только две детали  
в) хотя бы одна деталь  
г) от одной до двух деталей?

```python
# Вероятности выхода из строя деталей
p1 = 0.1
p2 = 0.2
p3 = 0.25

# Вероятность выхода из строя всех деталей
p_fail_all = p1 * p2 * p3

# Вероятность выхода из строя только двух деталей
p_fail_two = ((1-p1) * p2 * p3) + (p1 * (1-p2) * p3) + (p1 * p2 * (1-p3))

# Вероятность того, что хотя бы одна деталь выйдет из строя
p_fail_at_least_one = 1 - ((1-p1) * (1-p2) * (1-p3))

# Вероятность того, что выйдет из строя от одной до двух деталей
p_fail_one_or_two = p_fail_two + (p1 * (1-p2) * (1-p3)) + ((1-p1) * p2 * (1-p3)) + ((1-p1) * (1-p2) * p3)

print('Вероятность того, что в первый месяц выйдут из строя:')
print(f'а) все детали {p_fail_all * 100 : .2f}')
print(f'б) только две детали {p_fail_two * 100 : .2f}')
print(f'в) хотя бы одна деталь {p_fail_at_least_one * 100 : .2f}')
print(f'г) от одной до двух деталей {p_fail_one_or_two * 100 : .2f}')
```
